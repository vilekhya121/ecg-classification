# DATA PREPROCESSING

1.PERFORMING LABEL ENCODING ON CLASSES


```python
# LABEL ENCODING THE VARIABLES

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data.iloc[:,-1] = le.fit_transform(data.iloc[:,-1])
data
```

# SPLITTING THE DATASET

1. SPLITTING THE DATA SET INTO 'X' VARIABLE AND 'y' VARIABLE
2. PERFORMING THE SMOTE-ENN OPERATION FOR BALANCING THE CLASSES
3. VISUALIZING THE RESAMPLED CLASSES

```python
# SPLITING THE DATA INTO FEATURES (X) AND LABEL (y)

# X includes all the column values except the last column(Label column)

X = data.iloc[:,:-1].values
X
```

```python
# y represents the last column which is the labels columns
# No need of labelling as the data is already labeled

y = data.iloc[:,-1].values
y
```

```python
# SPLITTING THE DATA INTO TRAINING AND TESTING DATASETS
# IMPORTING TRAIN TEST SPLIT FROM SKLEARN MODEL SELECTION

from sklearn.model_selection import train_test_split

from imblearn.combine import SMOTEENN
smote_enn = SMOTEENN()

X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2,random_state = 0)

# APPLYING THE SMOTE-ENN TO BALANCE THE CLASSES ASWELL AS REMOVE THE NOISE
X_resampled, y_resampled = smote_enn.fit_resample(X_train, y_train)


print("The X_train:")
print(X_resampled)
print("")
print("The X_test:")
print(X_test)
print("")
print("The y_train:")
print(y_resampled)
print("")
print("The y_test:")
print(X_test)
print("")
```

```python
plt.figure(figsize = (10,8))
plt.hist(y_resampled)
plt.xlabel("Classes")
plt.ylabel("Count of the Classes")
plt.title("The Count Of The Classes")
plt.show()
```

# MODEL BUILDING 

1. BUILDING THE RANDOMFOREST CLASSIFIER MODEL
2. TRAINING THE MODEL
3. MAKING THE PREDICTION

```python
# MODEL BUILDING AND TRAINING THE MODEL
# IMPORTING RANDOM FOREST CLASSIFIER

from sklearn.ensemble import RandomForestClassifier

# RANDOM FOREST CLASSIFIER
rf_classifier = RandomForestClassifier()

# TRAINING THE CLASSIFIER
rf_classifier.fit(X_resampled,y_resampled)

# PREDICTING USING THE TRAINED CLASSIFIER
y_pred = rf_classifier.predict(X_test)

print("Predicted 'y' values:")
print(y_pred)
print("Actual 'y' values:")
print(y_test)
```

# PERFORMANCE EVALUATION

1.ACCURACY SCORE
2.CLASSIFICATION REPORT 
3.CONFUSION MATRIX

```python
# EVALUATING THE PERFORMANCE OF THE CLASSIFIER
# IMPORTING METRICS FOR EVALUATION

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# ACCURACY SCORE
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("")

# CLASSIFICATION REPORT
classification_report = classification_report(y_test,y_pred,target_names=labels)
print("Classification Report:")
print(classification_report)
print("")

# CONFUSION MATRIX
confusion_matrix = confusion_matrix(y_test,y_pred)

# HEATMAP OF CONFUSION MATRIX
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix/np.sum(confusion_matrix), annot=True, fmt='.2%', cmap='Blues', xticklabels=labels ,yticklabels=labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

# CROSS VALIDATION 
1.CREATING THE MODEL FOR STANDARDIZATION AND CLASSIFICATION
2.CREATING THE CROSS VALIDATION STRATEGY
3.PERFORMING THE 5-FOLD CROSS VALIDATION
4.EVALUATING THE CROSS VALIDATION RESULT

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.pipeline import make_pipeline


classifier = RandomForestClassifier()

# Creating a pipeline with standardization
model = make_pipeline(classifier)

# Cross-validation strategy
cv = StratifiedKFold(n_splits=5, shuffle=True)

# 5-fold cross-validation
cross_val_results = cross_val_score(model, X_resampled, y_resampled, cv=cv, scoring='accuracy')

# Print the cross-validation results
print("Cross-Validation Results:")
print(cross_val_results)
print("Mean Accuracy:", cross_val_results.mean())
```

```python
accuracy_variance = np.var(cross_val_results)
print("Variance of Accuracy:", accuracy_variance)
```

The model achieved very high accuracy, consistently around 99.8%, on all folds of the cross-validation. This suggests that the model is performing well on the given dataset.
The variance of accuracy is extremely small (3.22e-08), indicating very little variability in the model's performance across different folds. This suggests that the model's performance is consistent.
